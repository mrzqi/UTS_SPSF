```python
# For f(x)=sinx and 0<=x<=pi make code to calculate area below the curve using rectangle left point, rectangle mid point, rectangle right point, trapezium, and Simpson method for 10 partition. Explain which method gives the best result. What is the analytical solution of the problem? Name each method with different function, e.g rectl(), rectm(), rectr(), trapez(), and simpson() with xbeg, xend, partition as parameters.

import math

def f(x):
    return math.sin(x)

def rectl(xbeg, xend, partition):
    dx = (xend - xbeg) / partition
    area = 0
    for i in range(partition):
        x = xbeg + i*dx
        area += f(x)*dx
    return area

def rectm(xbeg, xend, partition):
    dx = (xend - xbeg) / partition
    area = 0
    for i in range(partition):
        x = xbeg + i*dx + dx/2
        area += f(x)*dx
    return area

def rectr(xbeg, xend, partition):
    dx = (xend - xbeg) / partition
    area = 0
    for i in range(partition):
        x = xbeg + i*dx + dx
        area += f(x)*dx
    return area

def trapez(xbeg, xend, partition):
    dx = (xend - xbeg) / partition
    area = 0
    for i in range(partition):
        x1 = xbeg + i*dx
        x2 = x1 + dx
        area += (f(x1) + f(x2)) * dx / 2
    return area

def simpson(xbeg, xend, partition):
    dx = (xend - xbeg) / partition
    area = 0
    for i in range(partition):
        x1 = xbeg + i*dx
        x2 = x1 + dx
        xmid = (x1 + x2) / 2
        area += (f(x1) + 4*f(xmid) + f(x2)) * dx / 6
    return area

xbeg = 0
xend = math.pi
partition = 10

area_rectl = rectl(xbeg, xend, partition)
area_rectm = rectm(xbeg, xend, partition)
area_rectr = rectr(xbeg, xend, partition)
area_trapez = trapez(xbeg, xend, partition)
area_simpson = simpson(xbeg, xend, partition)

print("Luas dengan rectangle left point: ", area_rectl)
print("Luas dengan rectangle mid point: ", area_rectm)
print("Luas dengan rectangle right point: ", area_rectr)
print("Luas dengan trapezium: ", area_trapez)
print("Luas dengan Simpson method: ", area_simpson)
```

Dari hasil perhitungan di atas, dapat dilihat bahwa metode Simpson memberikan hasil yang paling akurat, diikuti oleh metode trapezium dan rectangle mid point.

```python
# Using Finite Difference Method (FDM) find plot the curve of vertical position againts time for a free fall motion if , ,  with number of grid by modifying a code. Explain the form of the curve. Which gives better result?

import numpy as np
import matplotlib.pyplot as plt

def free_fall(y0, yt, g, n):
    t0 = 0
    tt = np.sqrt(2*(yt-y0)/-g)
    t = np.linspace(t0, tt, n)
    dt = t[1]-t[0]
    y = np.zeros(n)
    v = np.zeros(n)
    y[0] = y0
    for i in range(1,n):
        y[i] = y[i-1] + v[i-1]*dt
        v[i] = v[i-1] + g*dt
    return t, y

y0 = 10
yt = 20
g = -10
n = [10, 20, 40, 80, 160]

for i in n:
    t, y = free_fall(y0, yt, g, i)
    plt.plot(t, y, label='n = {}'.format(i))

plt.title('Kurva posisi vertikal pada gerak jatuh bebas')
plt.xlabel('Waktu (s)')
plt.ylabel('Posisi vertikal (m)')
plt.legend()
plt.show()
```

Semakin besar jumlah grid waktu (n), semakin halus kurva posisi vertikalnya dan semakin akurat.
n = 160 memberikan hasil yang lebih baik dibandingkan n yang lebih kecil karena kurva posisi vertikalnya lebih halus dan lebih dekat dengan hasil analitisnya.jika dibandingkan antara n=10, 20, 40, 80, dan 160, n=160 memberikan hasil yang lebih baik karena kurva yang dihasilkan semakin halus dan mendekati solusi eksak.
